```python
#1. Countdown
#Start from the value of x and countdown by 1 to zero

def countdown(x):
    myList = [i for i in range(x,-1,-1)]
    return (myList)
start = 10
print(countdown(start))
```

```python
#2. Print and Return
#A list argument is passed to function's single parameter, and function prints the position 0 item
#and position 1 item is returned

def print_and_return(myList):
    print (myList[0])
    return (myList[1])

yourList = [1,2]
print_and_return(yourList)
```

```python
#3. First Plus Length
#A list argument is passed to function's single parameter, function adds value of position 0 to the length of list argument

def first_plus_length(myList):
    return(myList[0] + len(myList))

yourList = [5,2,3,4,5,3]
print(first_plus_length(yourList))
```

```python
#4. This Length, That Value
#The function accepts two arguments and the function returns a list;
#length of list is first argument, each and all list items are the second argument

def length_and_value(x,y):
    return ([y for i in range(x)])

print(length_and_value(4,7))
print(length_and_value(6,2))
```

```python
#5. Values Greater than Second
#The function accepts a list argument.
#The function uses the item in position 1 of the list to create a new list (sub items of passed list argument aka original list)
#The new list is based on only items that are greater the position 1 item in the passed list argument

def values_greater_than_second(myList):
    if len(myList) > 1:
        newList = [i for i in myList if i > myList[1]]
        return(newList)
    else:
        return (False)
    print(newList)

yourList = [5,2,3,2,1,4]
print(values_greater_than_second(yourList))

yourList = [2]
print(values_greater_than_second(yourList))
```
