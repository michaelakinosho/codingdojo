```python
#1. Biggie Size
#Function accepts a list argument, and replaces any item in list that is positive i.e. greater than zero

def biggie_size(myList):
    myList = ["big" if i > 0 else i for i in myList]
    #print(myList)
    return (myList)

yourList = [-1,3,5,-5]
biggie_size(yourList)
```

```python
#2. Count Positives
#Function accepts a list argument, and replaces the last item in the list with count of positive numbers in the list
#Function also needs to return the updated list

def count_positives(myList):
    
    myList[-1] = len([i for i in myList if i > 0])
    return (myList)

yourList = [-1,1,1,1]
print(count_positives(yourList))

yourList = [1,6,-4,-2,-7,-2]
print(count_positives(yourList))
```

```python
#3. Sum Total
#Function accepts a list argumnet, and returns the sum of the up all the items in the list

def sum_total(myList):
    total = 0
    for i in myList:
        total += i
    return (total)

yourList = [1,2,3,4]
sum_total(yourList)
```

```python
#4. Average
#Function accepts a list argument, and returns the average of the list items

def average(myList):
    total = 0
    count = 0
    for i in myList:
        total += i
        count += 1
    return (total/count)

yourList = [1,2,3,4]
average(yourList)
```

```python
#5. Length

def length(myList):
    count = 0
    for i in myList:
        count += 1
    return (count)

yourList = [37,2,1,-9]
print(length(yourList))

yourList = []
print(length(yourList))
```

```python
#6. Minimum

def minimum(myList):
    count = 0
    for i in myList:
        count += 1
    
    if count == 0:
        return (False)
    else:
        min_num = 0
        for i in myList:
            if i < min_num:
                min_num = i
        return (min_num)

yourList = [37,2,1,-9]
print(minimum(yourList))

yourList = []
print(minimum(yourList))
```

```python
#7. Maximum

def maximum(myList):
    count = 0
    for i in myList:
        count += 1

    if count == 0:
        return (False)
    else:
        max_num = 0
        for i in myList:
            if i > max_num:
                max_num = i
        return (max_num)

yourList = [37,2,1,-9]
print(maximum(yourList))

yourList = []
print(maximum(yourList))
```

```python
#8. Ultimate Analysis

def ultimate_analysis(myList):
    count = 0
    for i in myList:
        count += 1
    
    if count < 1:
        return (False)

    myDict = {'sumTotal':0,'average':0,'minimum':0,'maximum':0,'length':0}
    for i in myList:
        myDict['sumTotal'] += i
        myDict['length'] += 1
        myDict['average'] = myDict['sumTotal']/myDict['length']
        if i < myDict['minimum']:
            myDict['minimum'] = i
        
        if i > myDict['maximum']:
            myDict['maximum'] = i
    return (myDict)

yourList = [37,2,1,-9]
print(ultimate_analysis(yourList))

yourList = []
print(ultimate_analysis(yourList))
```

```python
#9. Reverse List

def reverse_list(myList):
    temp = 0
    count = 0
    for i in myList:
        count += 1

    i = 0
    while i < count/2:
        temp = myList[i]
        myList[i] = myList[count-1-i]
        myList[count-1-i] = temp
        i += 1

    return (myList)

yourList = [37,2,1,-9]
print(reverse_list(yourList))

yourList = [37,2,1]
print(reverse_list(yourList))

yourList = [37,2,1,34,-9,56]
print(reverse_list(yourList))
```
