```python
#Task 3) Determine the amount of money collected for each item
print('Array of prices for each item: ',price_array)
print('Array of quantities sold for each item: ',quantity_sold_array)
money_collected_array = price_array * quantity_sold_array
print('Array of money collected for each item: ',money_collected_array)
print('Total amount of money collected: ',money_collected_array.sum())
```

```python
#Task 4) Create a Numpy array of the quantities remaining
#Adding zero for the sold out items ensures that bake club arrays have the same dimension size/length across all arrays
quantity_leftover_array = np.array([0,0,2,1,6,0,2])
print('Array of quantities leftover for each item: ', quantity_leftover_array)
```

```python
#Task 5) Determine how much was collected from clearance sale; with all prices cut in half
price_array_clearance1 = price_array/2
print('Array of prices on clearance: ',price_array_clearance1)
money_collected_clearance1_array = price_array_clearance1 * quantity_leftover_array
print('Array of money collected from clearance: ', money_collected_clearance1_array)
print('Total amount of money collected from clearance', money_collected_clearance1_array.sum())
```

```python
#Task 6) Determine the grand total amount collected between the regular and clearance sales
grand_total_collected = money_collected_array+money_collected_clearance1_array
print('Grand total collected is: ', grand_total_collected.sum())
```
