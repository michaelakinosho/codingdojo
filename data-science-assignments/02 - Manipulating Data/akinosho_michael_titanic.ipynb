#2. Percentage of people who survived

```python
format(df['Survived'].mean(),".0%") + ' of the people survived'
```

```python
#Creating survived filter
survived_filter = df['Survived'] == 1
df.loc[survived_filter,:].info()
```

```python
df.loc[survived_filter==False,:].shape
```

#3. The number of each sex that survived

```python
df.loc[survived_filter,'Sex'].value_counts()
```

```python
#Creating filter for paying less than 10 for fare
less_than_10_fare_filter = df['Fare'] < 10
df.loc[less_than_10_fare_filter,:].info()
```

#4. Percentage of group of passengers paying less than 10 in fare that survived

```python
format(df.loc[survived_filter & less_than_10_fare_filter,'Fare'].count()/df.loc[survived_filter,'Fare'].count(), "0.0%") + ' represents the percentage that survived and paid less than 10 for their fare'
```

```python
df.loc[less_than_10_fare_filter,['Sex','Survived']].value_counts(normalize=True).sum('Survived'=='0')
```

#5. Average age of those who didn't survive

```python
#Choose not to replace the rows with an NaN value with the average age of their respective sex
#Some side analysis showed that age was not a driving factor leading/contributing to survival
format(df.loc[survived_filter==False,'Age'].mean(),",.0f") + " years old, average of people who did not survive"
```

#6. Average age of those who did survive

```python
format(df.loc[survived_filter==True,'Age'].mean(),",.0f") + " years old, average of people who survived"
```

#7. Average age of those who survived and didn't grouped by Sex

```python
df.groupby(['Sex','Survived'])[['Age']].mean()
```
