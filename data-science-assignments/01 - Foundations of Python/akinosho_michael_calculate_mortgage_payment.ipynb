```python


#calculate monthly payment
def cal_payment(P,int_rate,years):
    n = years * 12
    r = int_rate/12

    ans = P * ((r*(1+r)**n)/((1+r)**n - 1))
    print(round(ans,0))

a = 300000      #amount borrowed
i = 0.01 #annual interest in decimal not percent
y = 30      #number of years amount borrowed for
cal_payment(a,i,y)
```
