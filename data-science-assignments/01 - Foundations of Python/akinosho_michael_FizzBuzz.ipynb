```python
def checkFizzBuzz(start,end):
    for i in range(start,end+1):
        if i%3==0:
            if i%5==0:
                print("FizzBuzz")
            else:
                print("Fizz")
        elif i%5==0:
            print("Buzz")
        else:
            print(i)

startNum = 1 #integer value to start from inclusive
endNum = 35 #integer value to end at inclusive
checkFizzBuzz(startNum, endNum)
```
