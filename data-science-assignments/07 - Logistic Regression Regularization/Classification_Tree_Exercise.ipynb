```python
df.replace({'diagnosis':{'M':1,'B':0}},inplace=True)
```

```python
y = df['diagnosis']
X = df.drop(columns=['diagnosis'])
```

```python
num = 42
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = num)
```

```python
def run_classifiers(X_train, X_test, y_train, y_test, num):
    dec_tree = DecisionTreeClassifier(random_state=num)
    dec_tree.fit(X_train, y_train)

    bag_class = BaggingClassifier(random_state=num)
    bag_class.fit(X_train, y_train)
    
    rand_forest = RandomForestClassifier(random_state=num)
    rand_forest.fit(X_train, y_train)

    dec_tree_train = dec_tree.score(X_train, y_train)
    dec_tree_test = dec_tree.score(X_test, y_test)
    print("Decision Tree Classifier Train Score: ", dec_tree_train)
    print("Decision Tree Classifier Test Score: ", dec_tree_test)

    bag_class_train = bag_class.score(X_train, y_train)
    bag_class_test = bag_class.score(X_test, y_test)
    print("Bagging Classifier Train Score:", bag_class_train)
    print("Bagging Classifier Test Score:", bag_class_test)

    rand_forest_train = rand_forest.score(X_train, y_train)
    rand_forest_test = rand_forest.score(X_test, y_test)
    print("Random Forest Classifier Train Score:", rand_forest_train)
    print("Random Forest Classifier Test Score:", rand_forest_test)
```

```python
run_classifiers(X_train, X_test, y_train, y_test, num)
```

Recommendation is the random forest classifier
