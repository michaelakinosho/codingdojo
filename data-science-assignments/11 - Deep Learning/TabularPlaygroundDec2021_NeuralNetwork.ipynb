```python
num_selector = make_column_selector(dtype_include='number')
```

```python
scaler = StandardScaler()
```

```python
scaler_tuple = (scaler, num_selector)
column_transformer = make_column_transformer(scaler_tuple)
```

```python
param_grid = {
    'n_estimators': [150],
    'num_leaves': [20],
    'max_depth': [2],
    'learning_rate':[0.1],
    'random_state':[42],
    'class_weight':['balanced']
}
```

```python
myBoostClassifier = LGBMClassifier()
```

```python
X_test = column_transformer.transform(X_test)
```

```python
X_train = column_transformer.transform(X_train)
```

```python
pipe = GridSearchCV(estimator=myBoostClassifier,param_grid=param_grid)
```

```python
pipe.fit(X_train, y_train)
```

```python
predictions = pipe.predict(X_test)
```

```python
df_submission['Cover_Type'] = pd.Series(predictions)
```
