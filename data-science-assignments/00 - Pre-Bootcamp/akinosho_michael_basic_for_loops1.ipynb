```python
#3. Counting, the Dojo Way
# Print integers 1 to 100
# If divisible by 10 print Coding Dojo
# If divisible by 5 print Coding, else print the integer
start = 5
stop = 100
for i in range(start,stop+1):
    if i%10==0:
        print("Coding Dojo")
    elif i%5==0:
        print("Coding")
    else:
        print(i)
```

```python
#5. Countdown by Fours
#Print positive numbers starting from 2018 and counting down to zero
start = 2018
stop = 0
interval = -4
myList = [i for i in range(start, stop, interval)]

for i in range(start,stop,interval):
    print(i)

while start > 0:
    print(start)
    start -=4

myList
```

```python
#6. Flexible Counter
#Print the multiples of the integer set to mult that are between
#variables lowNum and highNum
lowNum = 2
highNum = 9
mult = 3

# If mult is greater than lowNum set lowNum to mult
if mult >= lowNum:
    lowNum = mult

while lowNum < highNum + 1:
    #If statement tries to increment lowNum by 1 until first multiple is reached if mult is smaller than lowNum
    if lowNum % mult > 0:
        lowNum += 1
    #Start printing the multiples by simply incrementing from the first multiple identified
    else:
        print(lowNum)
        lowNum += mult
    
    
```
