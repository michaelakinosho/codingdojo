```python
import random

myList = [random.randint(0,5000000) for i in range(0,100)]
print(myList)

def insertion_sort(myList):
    #This loop starts the sort at position 1, comparing to position 0, to see if pos 1 value is less than pos 0
    for i in range(1,len(myList)):
        temp = i
        #Needed to add this list comprehension line, trying to figure out the sub list to the left was proving to be very mentally challenging and unmanageable
        #It really helped in doing the comparisons
        newList = [j for j in myList[0:i+1]]

        #This loop does the actual heavylifting, starting from the end of the list and working backwards to pos 0
        #Checking at pos -1 if i above is less than each item in newList
        #Once item myList[i] is greater than item newList[in the current iteration] this pos is length of newList minus j and minus 1
        #This allows movement from right to left
        for j in range(len(newList)):
            if myList[i] < newList[len(newList)-j-1]:
                temp = len(newList)-j-1

        if temp != i:
            myList.insert(temp,myList.pop(i))

insertion_sort(myList)
print('\n', myList)
```
