```python
import random

myList = [random.randint(0,10000) for i in range(0,10)]
print(myList)

def selection_sort(myList):
    #This outer for Loop ensures iteration based on the length of the list
    for x in range(len(myList)):
        myNum = myList[x]
        
        #At first glance idx = x appears useless or not even needed, but without it the pop and insert actually does not work
        idx = x
        
        #This inner for Loop checks for items smaller than myNum and if smaller items are found keeps track of them
        for y in myList[x+1:]:
            if y < myNum:
                myNum = y
                idx = myList.index(y)

        #This pops the smaller item found in the sub list and inserts it in position provided in x in the outer Loop up above
        #x indicates where in the list the insertion needs to happen, while idx indicates which item to remove and insert in position x
        myList.insert(x,myList.pop(idx))

selection_sort(myList)
myList
```
