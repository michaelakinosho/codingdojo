```python
class User:
    def __init__(self, username, email_address):
        self.name = username
        self.email = email_address
        self.account_balance = 0
    
    def make_deposit(self, amount):
        self.account_balance += amount

michael = User("Guido van Rossum", "guido@python.com")
anna = User("Monty Python","monty@python.com")

print(michael.name)
print(anna.name)
michael.make_deposit(1000)
michael.account_balance
```
