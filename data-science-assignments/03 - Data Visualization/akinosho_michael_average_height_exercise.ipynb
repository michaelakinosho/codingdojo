#1c. Which columns have missing values?

```python
print('Columns with missing values are:')
for i in df.columns[df.isna().any()]:
    print(i)
print("A high percentage of Olympic athletes don't earn a medal")
```

```python
city_name = 'London'
city_filter = df['City'] == city_name
res = df.loc[city_filter,'City'].value_counts()
```

#1d. How many entries correspond to the city of London?

```python
print('Entries corresponding to',city_name,'are', format(res[0],","))
```

#1e. What is age of the youngest athlete?

```python
print('Age of youngest athlete is',df['Age'].min())
```

#2a. How many athletes who participated in the Tug-Of-War were from Sweden?

```python
Team_filter = df['Team'] == 'Sweden'
Sport_filter = df['Sport'] == 'Tug-Of-War'

res = df.loc[Team_filter & Sport_filter,['Team','Sport','ID']].value_counts()
```

```python
res
```

```python
ans = res.sum()
print('Number of athletes in Tug-Of-war from Sweden:',ans)
```

#2b. How many Gold medals were awarded to athletes over 40 years old?

```python
Medal_filter = df['Medal'] == 'Gold'
Age_filter = df['Age'] > 40.0

res = df.loc[Medal_filter & Age_filter, ['ID']].value_counts()
```

```python
ans = res.sum()
print('Number of athletes over 40 years old, that earned a gold medal:', ans)
```
