```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
plt.hist([df['Percentage of Dog Owners']],label=['Dog'],color=['red'],edgecolor='black',bins=25);
plt.title('Displaying Histogram of Dog Owners')
plt.xlabel("Percentage of Dog Owners")
plt.ylabel('Count')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}%'))
plt.legend();
plt.show()
```

```python
plt.hist([df['Percentage of Cat Owners']],label=['Cat'],color=['yellow'],edgecolor='black',bins=25);
plt.title('Displaying Histogram of Cat Owners')
plt.xlabel("Percentage of Cat Owners")
plt.ylabel('Count')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}%'))
plt.legend();
plt.show()
```

#3a. The distribution of cat owners is concentrated around 25% to 40%, a spread of about 15%, while the distribution of dog owners is concentrated around 25% to 50%, a big spread of 25%. A higher percentage of pet owners have dogs.

```python
plt.hist([df['Percentage of Cat Owners'],df['Percentage of Dog Owners']],label=['Cat','Dog'],color=['yellow','red'],edgecolor='black',bins=15,alpha=0.7,width=1.5);
plt.title('Displaying Histogram of Animal Owners')
plt.xlabel("Percentage of Animal Owners")
plt.ylabel('Count')
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('{x:,.0f}%'))
plt.legend();
plt.show()
```

```python
cats = df['Mean Number of Cats']
dogs = df['Mean Number of Dogs per household']
fig, (reg_ax, notch_ax) = plt.subplots(nrows = 2,ncols = 1, figsize=(6,10))
labels = ['Cats','Dogs']

bplot1 = reg_ax.boxplot([cats,dogs],vert=True,patch_artist=True,labels=labels)
reg_ax.set_title('Mean Number of Cats and Dogs per Household')

bplot2 = notch_ax.boxplot([cats,dogs],vert=True,patch_artist=True,labels=labels,notch=True)
notch_ax.set_title('Mean Number of Cats and Dogs per Household')

colors = ['yellow','red']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

for ax in [reg_ax,notch_ax]:
    ax.yaxis.grid(True)
    ax.xaxis.grid(True)
    ax.set_xlabel('Pets')
    ax.set_ylabel('Mean Number of Pets')
    ax.set_title("Distribution of Mean Number of Pets per Household")

plt.show()
```

#5. The tendency is for a household to have a higher number of cats versus dogs. This is not saying that their are more cats than dogs owned.
